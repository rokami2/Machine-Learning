```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
#question 1
#data preprocessing

#Loading dataset
titanic = sns.load_dataset('titanic')

#dropping unneeded values and recoding sex and embarked
titanic = titanic.drop(['class', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone'], axis=1)
titanic['age'].fillna(titanic.groupby('pclass')['age'].transform("median"), inplace=True)
titanic.dropna(inplace=True)

new_sex = pd.get_dummies(titanic['sex'], drop_first=True)
new_embarked = pd.get_dummies(titanic['embarked'], drop_first=True)

titanic = pd.concat([titanic, new_sex, new_embarked], axis=1)
titanic.drop(['sex', 'embarked'],axis=1, inplace=True)

#show table of the data
titanic.head()
titanic.describe()
```

```python
#question 2
#create train/test split
# library to split dataset into train and test
from sklearn.model_selection import train_test_split

# X is all possible features/independent variables, Y is what we are modeling
X = titanic.drop('survived', axis=1)
Y = titanic['survived']

# split the data with 30% of rows going to the training set and a state seed of 1000 for consistency
trainX, testX, trainY, testY = train_test_split(X,Y,test_size=0.3, random_state = 1000)

# print the shape of the split dataframes
trainX.shape, trainY.shape, testX.shape, testY.shape

# started with 889 observations - 622 are for training and 267 for testing
```

```python
#question 3
from sklearn import tree

#make a tree with default options
treemax = tree.DecisionTreeClassifier()

#fit with data
treemax = treemax.fit(trainX, trainY)

print('the accuracy rate on training data is', treemax.score(trainX, trainY))

print('the accuracy rate on testing data is', treemax.score(testX, testY))

# create graph of decision tree
import graphviz
treemax_graph = tree.export_graphviz(treemax, out_file=None, 
            feature_names=['pclass', 'age', 'sibsp', 'parch', 'fare', 'male', 'Q', 'S'],
            class_names = ['did not survive', 'survived']  ,                        
            filled=True, rounded=True,  
            special_characters=True)  
graph = graphviz.Source(treemax_graph)  
graph
```

```python
#question 4 and 5
# remake the tree with a max depth of 2
treedepth = tree.DecisionTreeClassifier(max_depth = 2)

# fit with data
treedepth = treedepth.fit(trainX, trainY)

print('the accuracy rate on training data is', treedepth.score(trainX, trainY))

print('the accuracy rate on testing data is', treedepth.score(testX, testY))

# create graph of second decision tree
import graphviz
treedepth_graph = tree.export_graphviz(treedepth, out_file=None, 
            feature_names=['pclass', 'age', 'sibsp', 'parch', 'fare', 'male', 'Q', 'S'],
            class_names = ['did not survive', 'survived']  ,                        
            filled=True, rounded=True,  
            special_characters=True)
graph = graphviz.Source(treedepth_graph)
graph
```
