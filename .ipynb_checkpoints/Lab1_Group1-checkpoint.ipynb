```python
# import needed libraries #
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
```

```python
# import dataset from Seaborn 
# this link below provides more details about this dataset
# https://archive.ics.uci.edu/ml/datasets/Heart+failure+clinical+records

heartrecords = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
#figure 4a

#crosstab function, comparing values for patient survival based on the independent variable
sex_v_surv = pd.crosstab(index = heartrecords["DEATH_EVENT"], columns = heartrecords["sex"])

#label axes
sex_v_surv.index = ["Survived", "Deceased"]
sex_v_surv.columns = ["Female", "Male"]
sex_v_surv
```

```python
#figure 4b

sns.countplot(x = "DEATH_EVENT", hue = "sex", data = heartrecords)
```

```python
#figure 5a

smoke_v_surv = pd.crosstab(index = heartrecords["DEATH_EVENT"], columns = heartrecords["smoking"])
smoke_v_surv.index = ["Survived", "Deceased"]
smoke_v_surv.columns = ["Non-Smoking", "Smoker"]
smoke_v_surv
```

```python
#figure 5b

sns.countplot(x = "DEATH_EVENT", hue = "smoking", data = heartrecords)
```

```python
#figure 9

#create lists
zero = heartrecords['age'][heartrecords['DEATH_EVENT']==0].dropna()
one = heartrecords['age'][heartrecords['DEATH_EVENT']==1].dropna()

#draw two histograms 
plot.hist(zero, label = "Survived" )
plot.hist(one, label = "Fatality", color = "green")

#label the independent variable
plot.xlabel("age")

plot.legend(loc = "upper right")
plot.show()
```

```python
#figure 10

zero = heartrecords["creatinine_phosphokinase"][heartrecords["DEATH_EVENT"] == 0].dropna()
one = heartrecords["creatinine_phosphokinase"][heartrecords["DEATH_EVENT"] == 1].dropna()

plot.hist(zero, label = "Survived" )
plot.hist(one, label = "Fatality", color = "green")
plot.xlabel("Creatinine Phosphokinase")

plot.legend(loc = "upper right")
plot.show()
```

```python
#figure 15

#obtain correlation matrix for the values in the array
corr_mat = heartrecords.corr().round(2)

#customize the size of the plot
plot.figure(figsize = (10, 7))

#draw an annotated heatmap with colorscheme crest, using the correlation matrix from above
sns.heatmap(corr_mat, annot = True, cmap = "crest")
```
