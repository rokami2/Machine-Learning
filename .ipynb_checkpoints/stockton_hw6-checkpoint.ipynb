```python
#import libraries
import numpy as np # numbers, arrays, numerical analysis
import matplotlib.pyplot as plt # plotting
import pandas as pd # data library
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib
```

```python
#read in the data
#--------------------------------

#read in file
food=pd.read_csv('fastfood.csv')

#check
food.describe(include='all')
```

```python
#check na
food.isna().sum()
```

```python
#clean the data

#replace missing entries with the average
food['fiber'].replace(np.nan, food['fiber'].median())
food['protein'].replace(np.nan, food['protein'].median())

#drop columns with many missing entries
food = food.drop(['vit_a', 'vit_c', 'calcium'], axis=1)

#drop unneeded columns
food = food.drop(['restaurant', 'item', 'salad'], axis=1)

#general row cleanup
food.dropna(inplace=True)

#encode restaurants to be numerical
#encoder = LabelEncoder()
#food['restaurant'] = encoder.fit_transform(food['restaurant'])

#check
food
```

```python
#correlation matrix

#obtain correlation matrix for the values in the array
corr_mat = food.corr().round(2)

#customize the size of the plot
plt.figure(figsize = (10, 7))

#draw an annotated heatmap with colorscheme crest, using the correlation matrix from above
sns.heatmap(corr_mat, annot = True, cmap = "seismic")
```

```python
#QUESTION 2
```

```python
#normalize the data
from sklearn.preprocessing import normalize

food = normalize(food, axis=0, norm='max')

food
```

```python
# create plot with number of components kept and proportion of variance explained
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

varianceArray = []
for k in range(1,food.shape[1]+1):
    pca_test = PCA(n_components = k)
    pca_test.fit(food)
    varianceArray.append([k,np.sum(pca_test.explained_variance_ratio_)])
    print("explained variance with ",k,"components: ",sum(pca_test.explained_variance_ratio_))
varianceArray = np.array(varianceArray)

plt.plot(varianceArray[:,0],varianceArray[:,1])
plt.xlabel('Number of Components Kept')
plt.ylabel('Variance Explained')
plt.ylim(0,1.05)
plt.xlim(1,4)
plt.show()
```

```python
#PCA with 2 and 5 components

#seperate PCA to grab the first 2 output components
pca2 = PCA(n_components = 2)
pca2.fit(food)

#seperate PCA to grab the first 2 output components
pca5 = PCA(n_components = 5)
pca5.fit(food)

# X2d is the first 2 principle components
X2d = pca2.transform(food)

# X5d is the first 5 principle components
X5d = pca5.transform(food)

print(X2d)
print(X5d)
```

```python
#correlation matrix

plt.matshow(pca5.components_)
plt.yticks([0,1, 2, 3, 4], ['1st Component', '2nd Component', '3rd Component', '4th Component', '5th Component'])
plt.colorbar()
plt.xticks(range(0,11), food.dtype.names, rotation = 60, ha = 'left')
plt.xlabel('Feature')
plt.ylabel('Principle Component')
```

```python
#visualize PCA output in 2d

plt.scatter(X2d[:,0], X2d[:,1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
#kmeans

#avoid memory leak
OMP_NUM_THREADS=2

#determine how many clusters with elbow test
wcss = [None]
kticks = [0]
for k in range(1,21):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(food)
    wcss.append(np.mean(kmeans.inertia_))
    kticks.append(k)
    
plt.plot(wcss)
plt.xlabel('K')
plt.ylabel('WCSS')
plt.xticks(kticks)
plt.xlim(0,20)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(food)
cluster_labels = kmeans.predict(food)

fig = plt.figure()
colors = ['green','blue','red','purple']
plt.scatter(food[:,0],food[:,1],c=cluster_labels,cmap=matplotlib.colors.ListedColormap(colors))
plt.axis('equal')

# to add cluster assignment labels to plot
ticks = np.unique(cluster_labels)
cb = plt.colorbar(spacing = 'proportional', ticks=ticks)
cb.set_label('corresponding cluster')

plt.show()
```
