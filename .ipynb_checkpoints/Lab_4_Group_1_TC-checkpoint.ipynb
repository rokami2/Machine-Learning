```python
#imports
from sklearn.datasets import load_iris
import seaborn as sns

#Load the iris dataset using seaborn
data = sns.load_dataset('iris')
#turn into an array
iris = data.values

print(iris)
```

```python
#remove 5th feature which is class name for X axis
X = iris[:,:-1]

#seperately grab the class name for Y axis
Y = iris[:,-1]

print(X)
```

```python
#normalize the data
from sklearn.preprocessing import normalize

X = normalize(X, axis=0, norm='max')

print(X)
```

```python
# create plot with number of components kept and proportion of variance explained
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

varianceArray = []
for k in range(1,X.shape[1]+1):
    pca_test = PCA(n_components = k)
    pca_test.fit(X)
    varianceArray.append([k,np.sum(pca_test.explained_variance_ratio_)])
    print("explained variance with ",k,"components: ",sum(pca_test.explained_variance_ratio_))
varianceArray = np.array(varianceArray)

plt.plot(varianceArray[:,0],varianceArray[:,1])
plt.xlabel('Number of Components Kept')
plt.ylabel('Variance Explained')
plt.ylim(0,1.05)
plt.xlim(1,4)
plt.show()
```

```python
#seperate PCA to grab the first 2 output components
pca = PCA(n_components = 2)
pca.fit(X)

# X2d is the first 2 principle components
X2d = pca.transform(X)

print(X2d)
```

```python
#plot all 3 classes with different colors using the 2 new components
plt.scatter(X2d[Y == 'virginica', 0], X2d[Y == 'virginica', 1], c='r', label='Virginica')
plt.scatter(X2d[Y == 'versicolor', 0], X2d[Y == 'versicolor', 1], c='b', label='Versicolor')
plt.scatter(X2d[Y == 'setosa', 0], X2d[Y == 'setosa', 1], c='g', label='Setosa')

labels = ['Virginica', 'Versicolor', 'Setosa']

plt.xlabel('PC1')
plt.ylabel('PC2')

plt.legend(labels)
```
